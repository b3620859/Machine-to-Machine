```python
import socket
import time
import mysql.connector
import datetime

HEADERSIZE = 10

s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
# s.bind((socket.gethostname(), 6688))
s.bind(('140.134.29.25', 6688))
s.listen(5)

BUFFER_SIZE = 1024  # Normally 1024, but we want fast response

connect = mysql.connector.connect(host="127.0.0.1", user="root", password="123", database="test2", auth_plugin='mysql_native_password')
if (connect):
    print("Connection Successful")
else:
    print("Connection Unsuccessful")

cursor = connect.cursor()

        
def get_date():
    unix = int(time.time())
    date = str(datetime.datetime.fromtimestamp(unix).strftime('%Y-%m-%d %H:%M:%S'))
    return date
    
    
def process_data_from_client(x):
    amb, obj = x.split(" ")
    return amb, obj


while True:
    # now our endpoint knows about the OTHER endpoint.
    clientsocket, address = s.accept()
    print(f"Connection from {address} has been established.")
    
    msg = "Welcome to the server!"
    msg = f"{len(msg):<{HEADERSIZE}}"+msg
    
    clientsocket.send(bytes(msg,"utf-8"))
    
    while True:
        recdata = clientsocket.recv(BUFFER_SIZE).decode('utf-8')
        amb, obj = process_data_from_client(recdata)
        date = get_date()
        print(date, amb, obj)
        
        # Write data to the database
        cursor.execute("INSERT INTO tempdata (Date, Amb_Temperature, Obj_Temperature) VALUES (%s, %s, %s)",(date, amb, obj)) # values has placeholders
        connect.commit() # SQL statement used for saving the changes
#         print(recdata)
        if not recdata:
            break
        else:
            pass
# close socket
s.close()
# close mysql
cursor.close
connect.close()
```
